```python
# Make sure you run this code cell everytime you open the notebook!
import networkx as nx            # Import NetworkX library as nx for its network functions
import numpy as np               # Import NumPy library as np for its mathematical functions
import matplotlib.pyplot as plt  # Import Matplotlib library as plt for its plotting functions

G=nx.Graph()        # Create the network
G.add_edges_from([("AbuKaahfa(P)","Wassi(P)"),("AbuKaahfa(P)","Zarar(P)"),("AbuKaahfa(P)","AbdulRehman(OTH)"),("Wassi(P)","Zarar(P)"),\
                  ("Wassi(P)","HafizArshad(LC)"),("Wassi(P)","AbuUmer(LC)"),("Wassi(P)","BabaImran(NH)"),("Wassi(P)","Nasir(NH)"),\
                  ("Zarar(P)","Fahadullah(OTH)"),("HafizArshad(LC)","Javed(TMH)"),("HafizArshad(LC)","AbuShoaib(TMH)"),\
                  ("HafizArshad(LC)","AbuUmer(LC)"),("Javed(TMH)","AbuUmer(LC)"),("Javed(TMH)","AbuShoaib(TMH)"),\
                  ("AbuShoaib(TMH)","AbuUmer(LC)"),("AbdulRehman(OTH)","Fahadullah(OTH)"),\
                  ("IsmailKhan(CST)","AjmalKasab(CST)"),("AjmalKasab(CST)","IsmailKhan(CST)")])

# A function defined for simple plotting
def plot_graph(G):
  pos = nx.drawing.layout.circular_layout(G)
  plt.figure(figsize=(15,15))                                                           # Specify the figure size
  nx.draw(G, pos=pos, with_labels=True, font_weight='bold', node_size=18000, \
        node_color='red', font_size=13, font_color='white',width=5, edge_color='green') # The figure properties
  plt.show()                                                                            # Show the network
  
  # Visualize the Graph
plot_graph(G)
```
