 ## Q1. (3 pts)

A simple network consists of N nodes in a single component. What is the maximum/minimum possible number of links it could have? 

Let's work out this example incrementally.

Let N=3. Draw the two networks (with the maximum number of links and minimum number of links) by hand.

Let N=4. Draw the two networks (with the maximum number of links and minimum number of links) and compare to N=3 case. 

Do you observe any trend? What is the maximum possible number of links a network with N nodes could have? What is the minimum possible number of links a network with N does could have?

## A1.



For N=3 we have a maximum of 3 links, and a minimum of 2 links.

For N=4 we have a maximum of 6 links, and a minimum of 3 links.

Please note that 3 is 3*(3-1)/2 (i.e., 3 choose 2), and 6 is 4*(4-1)/2 (i.e., 4 choose 2).

The maximum number of links is n*(n-1)/2 because there are [n choose 2 = n*(n-1)/2] distinct node pairs to put a link and each can have only one link in a simple network. 
The minimum number of links is n-1 since we require that the network be connected.

```python
#B

import networkx as nx            # Import NetworkX library for its network functions
import numpy as np               # Import NumPy library for its mathematical functions
import matplotlib.pyplot as plt  # Import Matplotlib library for its plotting functions


G = nx.Graph()                                                #Construct an undirected graph. 
G.add_edges_from([("Elisa", "Carl"), ("Carl", "Diana"), ("Carl", "Bob"), ("Anne", "Elisa"), ("Anne", "Diana"), ("Anne", "Bob"), ("Diana", "Bob")])    # Add the links. 

A = nx.to_numpy_matrix(G)                          # Create the adjacency matrix A. Note that this matrix is flipped from the one in class.
print("The adjacency matrix: \n", A)    # Print A 

print("The plot of the network: \n")                # Plot the network 
plt.figure(figsize=(6,4))                           # Set the figure size
nx.draw(G, with_labels=True, font_weight='bold', node_size=2500, \
        node_color='red', font_size=16, font_color='white',width=3, edge_color='green')  # Plot properties 
plt.show()                                          # Show the plot

print("Total number of links: ", G.number_of_edges())         # Print the number of links (L) 
print("Total degree for Bob:", G.degree[ "Bob"])              # Degree of node 4 
print("Average Degree: ", 2*G. number_of_edges ()/G. number_of_nodes()) # Average degree () = 2*L / N
```
