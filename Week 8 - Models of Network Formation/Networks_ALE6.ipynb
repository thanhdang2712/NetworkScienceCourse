```python
# ANSWER (CODE)
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edges = pd.read_csv('HIV_links.csv', skiprows = 1, sep=',', names=['Source', 'Target','Type'])
nodes = pd.read_csv('HIV_nodes.csv', skiprows=1, sep = ',', names = ['id', 'Label'])
nodes
```

```python
G = nx.from_pandas_edgelist(edges, 'Source', 'Target')
for i in sorted(G.nodes()):
    G.nodes[i]['Name'] = nodes.Label[i-1]
```

```python
G.nodes(data=True)
```

```python
plt.figure(figsize=(4,3))                                           # Specify the figure size
nx.draw(G, with_labels=True, font_weight='bold', node_size=500, \
        node_color='red', font_size=16, font_color='white',width=3, edge_color='green') # The figure properties
nx.draw_networkx_labels
plt.show()                                                          # Show the figure
```
