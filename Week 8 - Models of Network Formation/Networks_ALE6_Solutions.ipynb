```python
# ANSWER (CODE)
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edges = pd.read_csv('HIV_links.csv', skiprows = 1, sep=',', names=['Source', 'Target','Type'])
nodes = pd.read_csv('HIV_nodes.csv', skiprows=1, sep = ',', names = ['id', 'Label'])
```

```python
G = nx.from_pandas_edgelist(edges, 'Source', 'Target')
for i in sorted(G.nodes()):
    G.nodes[i]['Name'] = nodes.Label[i-1]
```

```python
# Function for Global Clustering Coefficient
from itertools import combinations

def global_clustering_coefficient(G):                         # A function for finding global clustering coefficient
    """
    Explicitly enumerate every open triangle.
    """
    G=G.to_undirected()
    open_triangle_nodes = []
    for node in list(G.nodes):
      neighbors = set(G.neighbors(node))
      for n1, n2 in combinations(neighbors, 2):
          if not G.has_edge(n1, n2):
              open_triangle_nodes.append([n1, node, n2])
    
    triangles=sum(nx.triangles(G).values())/3
    return 3*triangles/(3*triangles+len(open_triangle_nodes))
```

```python
# i. Centrality Analysis:
# NOTE!: You can always print out the 'Name' of patient i using G.nodes[i]['Name']
d_centralities = G.degree()                               # Find the degree centrality of all patients        
print('Sorted Degree Centrality Scores:', dict(sorted(d_centralities, key=lambda item: item[1], reverse=True)[0:9])) # Print the degree centrality of first 10 patients       
c_centralities=nx.closeness_centrality(G)                 # Find the closeness centrality of all patients
print('Sorted Closeness Centrality Scores:', dict(sorted(c_centralities.items(), key=lambda item: item[1], reverse=True)[0:10])) # Print the degree centrality of first 10 patients 
b_centralities = nx.betweenness_centrality(G)             # Find the harmonic centrality of all patients
print('Sorted Betweenness Centrality Scores:', dict(sorted(b_centralities.items(), key=lambda item: item[1], reverse=True)[0:10])) # Print the degree centrality of first 10 patients 
p_centralities = nx.pagerank(G,max_iter=100000)           # Find the closeness centrality of all patients
print('Sorted PageRank Centrality Scores:', dict(sorted(p_centralities.items(), key=lambda item: item[1], reverse=True)[0:10])) # Print the degree centrality of first 10 patients 
```

These observations lead me to conclude that Youtube recommend culinary contents based on most popular (most watched) videos based on all viewer's vote, leading it to recommend other videos from the same popular channels and trending topics regardless of their popularity. This bias in the algorithm makes sense as it takes advantage of the fact that most-watched videos by Youtube viewer populations as a whole maybe enjoyed to a random mainstream viewer.


However, I did notice that although Epicurious did overlap content and recommend videos from the other two channels, it was common that each channel recommended their own videos before recommending other food videos from other channels. This made the entirety of the network a weakly connected component rather than a strongly connected component. Overall, it was interesting to see how the topic of the videos that all 3 channels connected to was pasta. This says something about the eating trend in modern day and how pasta has become a very popular and largely eaten food around the world.
