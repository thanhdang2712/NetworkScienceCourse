## I. Game of Thrones

Game of Thrones was the hugely popular television series by HBO based on the book series <em>A Song of Ice and Fire</em> by George R.R. Martin. <p>The dataset constitutes a network for each book. These networks are given as text files describing the <em>links</em> between characters, with some attributes attached to each link. Let's start by loading in the networks.</p>






```python
# Make sure you run this code cell to import the relevant libraries and the networks!
import networkx as nx            # Import NetworkX library as nx for its network functions
import numpy as np               # Import NumPy library as np for its mathematical functions
import matplotlib.pyplot as plt  # Import Matplotlib library as plt for its plotting functions
import pandas as pd              # Import Pandas library as pd for its data manipulation and analysis functions

# The following code imports the networks from my Google Drive, and stores them to book1, book2, book3, book4, and book5 variables 
url1 = 'https://drive.google.com/file/d/1kiaWm1GrNlNfhJmQxo6qABTw9n8ddNOk/view?usp=sharing'
path1 = 'https://drive.google.com/uc?export=download&id='+url1.split('/')[-2]
book1 = pd.read_csv(path1)

url2 = 'https://drive.google.com/file/d/1LvkDY0xqBDyOF5CfbKe7EdkBogrSJmPe/view?usp=sharing'
path2 = 'https://drive.google.com/uc?export=download&id='+url2.split('/')[-2]
book2 = pd.read_csv(path2)

url3 = 'https://drive.google.com/file/d/17SXpzvjdJngtJYe5bD30YON6voN1NPvl/view?usp=sharing'
path3 = 'https://drive.google.com/uc?export=download&id='+url3.split('/')[-2]
book3 = pd.read_csv(path3)

url4 = 'https://drive.google.com/file/d/1Ygk0ORmfyNewepj41ErQRRzJp5TLvOdc/view?usp=sharing'
path4 = 'https://drive.google.com/uc?export=download&id='+url4.split('/')[-2]
book4 = pd.read_csv(path4)

url5 = 'https://drive.google.com/file/d/1QaDI3jhisxnQdMkj5QmCAxyrFgr-leHV/view?usp=sharing'
path5 = 'https://drive.google.com/uc?export=download&id='+url5.split('/')[-2]
book5 = pd.read_csv(path5)
```

## II. Network of Thrones
<p>The resulting DataFrames, for <code>book1</code> for example, has 5 columns: <code>Source</code>, <code>Target</code>, <code>Type</code>, <code>weight</code>, and <code>book</code>. Source and target are the two nodes that are linked. A network can have directed or undirected links and in this network all the links are undirected. The weight attribute of every link tells us the number of interactions that the characters have had over the book, and the book column tells us the book number.</p>
<p>Once we have the data loaded as a pandas DataFrame, it's time to create a network. We will use <code>Networkx</code> to create a network object for each book.</p>

```python
# Create a network object for each book. 
G_book1 = nx.Graph() 
G_book2 = nx.Graph()
G_book3 = nx.Graph()
G_book4 = nx.Graph()
G_book5 = nx.Graph()

# Currently, the network objects G_book1,..., G_book5 are empty. 
# Let's now populate it with the links from book1,..., book5. 
for _, edge in book1.iterrows():
    G_book1.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

for _, edge in book2.iterrows():
    G_book2.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

for _, edge in book3.iterrows():
    G_book3.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

for _, edge in book4.iterrows():
    G_book4.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

for _, edge in book5.iterrows():
    G_book5.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

# Creating a list of networks for all the books
books = [G_book1]
books.append(G_book2)
books.append(G_book3)
books.append(G_book4)
books.append(G_book5)
```

```python
# ANSWER (CODE):
def my_weighted_degree_centrality(WG):   # Weighted degree centrality function
  my_dict= dict(WG.degree(weight='weight'))
  total = sum(my_dict.values())
  result = {key: value / total for key, value in my_dict.items()}
  return result

d_centralities = my_weighted_degree_centrality(G_book5)             # Find the degree centrality of all characters in book 5
# A code piece that might help you to sort the centralities
print('Sorted Degree Centrality Scores:', dict(sorted(d_centralities.items(), key=lambda item: item[1], reverse=True)[0:9]))

c_centralities=nx.closeness_centrality(G_book5, distance='weight')  # Find the closeness centrality of all characters in book 5
# A code piece that might help you to sort the centralities
print('Sorted Closeness Centrality Scores:', dict(sorted(c_centralities.items(), key=lambda item: item[1], reverse=True)[0:9]))

# Please write a similar code for the other centrality metrics
```

## B. The evolution of character importance

According to degree centrality, the most important character in the first book is Eddard Stark but he is not even in the top 10 of the fifth book. The importance of characters changes over the course of five books because, you know, stuff happens... :)

Look at the evolution of centrality measures of a few important characters like ['Eddard-Stark', 'Tyrion-Lannister', 'Jon-Snow', 'Daenerys-Targaryen', 'Cersei-Lannister', 'Stannis-Baratheon', 'Sansa-Stark'] from first book to the fifth book.
```python
def my_weighted_degree_centrality(WG):  
  my_dict= dict(WG.degree(weight='weight'))
  total = sum(my_dict.values())
  result = {key: value / total for key, value in my_dict.items()}
  return result

evolution_of_characters = [my_weighted_degree_centrality(book) for book in books]    # Creating a list of degree centrality of all the books
degree_evolution_df = pd.DataFrame.from_records(evolution_of_characters)             # Creating a DataFrame from the list of degree centralities in all the books
# Table for the degree centrality evolution of some characters
print('DEGREE CENTRALITY EVOLUTION:\n', degree_evolution_df[['Eddard-Stark', 'Tyrion-Lannister', 'Jon-Snow', 'Daenerys-Targaryen', 'Cersei-Lannister', 'Stannis-Baratheon', 'Sansa-Stark']])

evolution_of_characters = [nx.closeness_centrality(book, distance='weight') for book in books] # Creating a list of closeness centrality of all the books
degree_evolution_df = pd.DataFrame.from_records(evolution_of_characters)                       # Creating a DataFrame from the list of closeness centralities in all the books 
# Table for the closeness centrality evolution of some characters
print('CLOSENESS CENTRALITY EVOLUTION:\n', degree_evolution_df[['Eddard-Stark', 'Tyrion-Lannister', 'Jon-Snow', 'Daenerys-Targaryen', 'Cersei-Lannister', 'Stannis-Baratheon', 'Sansa-Stark']])


evolution_of_characters = [nx.betweenness_centrality(book, weight='weight') for book in books] # Creating a list of betweenness centrality of all the books
degree_evolution_df = pd.DataFrame.from_records(evolution_of_characters)                       # Creating a DataFrame from the list of closeness centralities in all the books 
# Table for the closeness centrality evolution of some characters
print('BETWEENNESS CENTRALITY EVOLUTION:\n', degree_evolution_df[['Eddard-Stark', 'Tyrion-Lannister', 'Jon-Snow', 'Daenerys-Targaryen', 'Cersei-Lannister', 'Stannis-Baratheon', 'Sansa-Stark']])

evolution_of_characters = [nx.pagerank(book, weight='weight') for book in books] # Creating a list of pagerank centrality of all the books
degree_evolution_df = pd.DataFrame.from_records(evolution_of_characters)                       # Creating a DataFrame from the list of closeness centralities in all the books 
# Table for the closeness centrality evolution of some characters
print('PAGERANK CENTRALITY EVOLUTION:\n', degree_evolution_df[['Eddard-Stark', 'Tyrion-Lannister', 'Jon-Snow', 'Daenerys-Targaryen', 'Cersei-Lannister', 'Stannis-Baratheon', 'Sansa-Stark']])
```
